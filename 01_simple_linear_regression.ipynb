```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("Salary dataset.csv")
sns.jointplot(x='YearsExperience', y='Salary', data=df);
```

```python
plot = sns.regplot(x='YearsExperience', y='Salary', data=df)
plot.set_title("Linear Regression: Years of Experience vs Salary")
plot.set_xlabel("Years of Experience")
```

```python
from sklearn.linear_model import LinearRegression

X = pd.DataFrame({
    'YearsExperience': df['YearsExperience']
    })
y = df['Salary']

model = LinearRegression().fit(X, y)

while True:
  x = input("Provide YearsExperience (enter 'q' to exit): ")
  if x == 'q':
    break
  X_pred = pd.DataFrame({'YearsExperience': [float(x)]})
  predicted_salary = model.predict(X_pred)
  print(f"Predicted salary for {x} years of experience : {str(predicted_salary[0])}")
```
