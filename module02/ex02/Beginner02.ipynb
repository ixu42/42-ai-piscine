```python
# prepare the same data as in ex01
import pandas as pd

CarData = pd.read_csv('Car Price.csv')

CarData['CarCompany'] = CarData['CarName'].apply(lambda x: x.split(' ')[0])

corrections = {
    'alfa-romero': 'alfa-romeo',
    'maxda': 'mazda',
    'Nissan': 'nissan',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
}

CarData.CarCompany = CarData.CarCompany.replace(corrections)

CarData.CarCompany.unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# visualize 'CarComapny' variable against dependent variable 'price' via boxplot

# set figure size for better readability
plt.figure(figsize=(20, 10))

# create boxplot
sns.boxplot(
    data=CarData,
    x='CarCompany',
    y='price',
    hue='CarCompany',
    width=0.5,
    # define outlier properties
    flierprops=dict(marker='d', markersize=5, markerfacecolor='black')
)

plt.show()
```
