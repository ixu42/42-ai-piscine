```python
import pandas as pd

CarData = pd.read_csv('Car Price.csv')
```

```python
# extract the car's company's name from the variable 'CarName' into variable 'CarCompany'
CarData['CarCompany'] = CarData['CarName'].apply(lambda x: x.split(' ')[0])

CarData.head()
```

```python
# check the unique values under "CarCompany"
CarData['CarCompany'].unique()
```

```python
# replace the mis-spelling with correct ones

# dictionary for corrections
corrections = {
    'alfa-romero': 'alfa-romeo',
    'maxda': 'mazda',
    'Nissan': 'nissan',
    'porcshce': 'porsche',
    'toyouta': 'toyota',
    'vokswagen': 'volkswagen',
    'vw': 'volkswagen'
}

# replace misspelled names
CarData.CarCompany = CarData.CarCompany.replace(corrections)

# again checking the unique values
CarData.CarCompany.unique()
```
